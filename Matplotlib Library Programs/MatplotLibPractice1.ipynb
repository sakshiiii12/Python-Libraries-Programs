# What is Matplotlib?
- Matplotlib is a powerful Python library used to create static, interactive, and animated visualizations.
- It's especially useful for:
  1. Plotting graphs (line, bar, scatter, histogram, etc.)
  2. Customizing charts (colors, labels, titles, grid, legends)
  3. Creating subplots

```python
import matplotlib.pyplot as plt
```

# Basic Plots with Examples

```python
#1.Line Plot

x = [1,2,3,4,5]
y = [10,12,8,14,11]

plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

```python
#2.Bar Chart

categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 99]

plt.bar(categories, values, color='pink')
plt.title("Bar Chart")
plt.show()
```

```python
#3.Historgram

data = [1, 2, 2, 3, 3, 3, 4, 4, 5]

plt.hist(data, bins=5, color='lightgreen')
plt.title("Histogram")
plt.show()
```

```python
#4.Scatter plot

x = [5, 7, 8, 7, 2, 17]
y = [99, 86, 87, 88, 100, 86]

plt.scatter(x, y, color='blue')
plt.title("Scatter Plot")
plt.show()
```

```python
#5.Subplots(Multiple Charts)

plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [4, 5, 6])

plt.subplot(1, 2, 2)
plt.bar([1, 2, 3], [4, 5, 6])

plt.tight_layout()
plt.show()
```
