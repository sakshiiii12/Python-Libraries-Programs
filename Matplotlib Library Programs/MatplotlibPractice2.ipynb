```python
import matplotlib.pyplot as plt
```

```python
#Exercise 1: Line plot
#Plot the growth of a plant over 5 days.

days = [1, 2, 3, 4, 5]
height = [2, 3.5, 5, 7, 10]

plt.plot(days, height, color='red', marker='o')
plt.title("Growth of a Plant")
plt.xlabel("Days")
plt.ylabel("Height")
plt.grid(True)
plt.show()
```

```python
#Exercise 2: Bar Chart
#Compare the number of students in 4 different courses.

courses = ['Python', 'Java', 'Data Science', 'AI']
students = [150, 120, 180, 90]

plt.bar(courses, students, color='pink')
plt.title("Students Marks")
plt.xlabel("Courses")
plt.ylabel("Students Marks")
plt.show()
```

```python
#Exercise 3:Histogram
#Plot the distribution of marks for a class.

marks = [45, 67, 89, 70, 56, 90, 82, 75, 60, 50, 45, 80]

plt.hist(marks, color='lightblue', bins=5, edgecolor='black')
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.title("Marks of a class")
plt.grid(True)
plt.show()
```

```python
#Exercise 4:Scatter plot
#Visualize the relationship between hours studied and marks obtained.

hours_studied = [1, 2, 3, 4, 5, 6, 7, 8]
marks = [50, 55, 60, 65, 70, 78, 85, 90]

plt.scatter(hours_studied, marks, color='red', marker='o')
plt.title("Relationship between hours studied and marks obtained")
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
#Exercise 5: Subplots
#Plot the following:
#- Line plot of height growth
#- Bar chart of student count

# Line Plot Data: Plant height over days
days = [1, 2, 3, 4, 5]
height = [2, 3.5, 5, 7, 10]

# Bar Chart Data: Student counts
courses = ['Python', 'Java', 'Data Science', 'AI']
students = [150, 120, 180, 90]

#Line plot
plt.subplot(1, 2, 1)
plt.plot(days, height, marker='o', color='green')
plt.title("Plant Height Over Days")
plt.xlabel("Days")
plt.ylabel("Height (cm)")
plt.grid(True)

#Bar Chart
plt.subplot(1, 2, 2)
plt.bar(courses, students, color='skyblue')
plt.title("Students in Each Course")
plt.xlabel("Courses")
plt.ylabel("Number of Students")
plt.tight_layout()
plt.show()
```
