# What is Seaborn?
- Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-
  level interface for drawing attractive and informative statistical graphics.
# Why Use Seaborn?
- Simplifies complex visualizations (like heatmaps, violin plots, etc.)
- Integrates well with Pandas DataFrames
- Offers beautiful, default themes and color palettes
- Supports statistical plotting like regression lines, distribution plots, etc.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

2.Basic Plots
- Plot a barplot of average tip for each day.
- Use a countplot to show the number of males and females in the dataset.

```python
sns.barplot(x="day", y="tip", data=df)
plt.show()
```

```python
sns.countplot(x="sex", data=df)
plt.show()
```

3.Styling Practice
- Use the "whitegrid" style and "pastel" palette.
- Show a boxplot of total_bill by sex.

```python
sns.set_style("whitegrid")
sns.boxplot(x="sex", y="total_bill", data=df, palette='pastel')
plt.show()
```

4.Distribution Plot
- Plot the distribution of total_bill using sns.histplot() and sns.kdeplot().

```python
sns.histplot(df["total_bill"], kde=True )
plt.show()
```

```python
sns.kdeplot(df["total_bill"], fill=True)
plt.show()
```

5.Scatter Plot + Hue
- Use sns.scatterplot() to visualize total_bill vs tip, color-coded by time (lunch/dinner).

```python
sns.scatterplot(x="total_bill", y="tip", hue="time", data=df)
plt.show()
```

6.Box vs Violin
- Create both a boxplot and a violinplot of tip by smoker status. What differences do you see?

```python
sns.boxplot(x="smoker", y="tip", data=df)
plt.show()
```

```python
sns.violinplot(x="smoker", y="tip", data=df)
plt.show()
```

7.Linear Regression
- Use sns.lmplot() to draw a regression line for total_bill vs tip.
- Use hue='sex' and col='time' to split by subgroups.

```python
sns.lmplot(x="total_bill", y="tip", data=df, hue="sex", col="time")
plt.show()
```

10.Categorical Plot Grid
- Use sns.catplot() to plot tip by day, with different plots (kind="box", "violin", "strip").
- Use col='sex' or row='smoker' to add subplots.

```python
sns.catplot(x="day", y="tip", kind="box", col="sex", data=df)
plt.show()
```

```python
sns.catplot(x="day", y="tip", kind="violin", row="smoker", data=df)
plt.show()
```

```python
sns.catplot(x="day", y="tip", kind="strip", col="sex", row="smoker", data=df)
plt.show()
```

11.FacetGrid Custom Plotting
- Use sns.FacetGrid() on the diamonds dataset (load via sns.load_dataset('diamonds')).
- Create a grid with row='cut' and col='color', and map sns.histplot() on price.

```python
df4 = sns.load_dataset('diamonds')
```

```python
df4
```

```python
g = sns.FacetGrid(df4, row="cut", col="color", margin_titles=True)
g.map(sns.histplot, "price")
plt.show()
```

12.Heatmap Challenge
- Create a custom 10Ã—10 DataFrame with random numbers.
- Plot a heatmap with a custom color palette (e.g., "coolwarm").
- Add lines or annotations to highlight cells above a certain value.

```python
data = np.random.rand(10, 10)
df5 = pd.DataFrame(data, columns=[f"C{i}" for i in range(10)])
sns.heatmap(df5, cmap="magma", annot=True, linewidths=0.5)
plt.show()
```

13.Create a dashboard-style plot combining scatter, histogram, and boxplot using matplotlib subplots and Seaborn visualizations.

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

sns.scatterplot(x="total_bill", y="tip", data=df, ax=axs[0])
axs[0].set_title("Scatter Plot")

sns.histplot(df["total_bill"], kde=True, ax=axs[1])
axs[1].set_title("Histogram")

sns.boxplot(y="total_bill", data=df, ax=axs[2])
axs[2].set_title("Box Plot")

plt.tight_layout()
plt.show()
```
