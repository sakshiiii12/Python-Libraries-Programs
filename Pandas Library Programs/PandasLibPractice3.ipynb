13. Apply Custom Functions
Write a custom function to assign grades (A, B, C, etc.) based on Score and apply it to the DataFrame.

```python
def assign_grade(score):
    if score >= 90: return 'A'
    elif score >= 75: return 'B'
    elif score >= 60: return 'C'
    elif score >= 50: return 'D'
    else: return 'F'

df['Grade'] = df['Score'].apply(assign_grade)
```

```python
df
```

15. Resampling Time Series Data
Given a DataFrame with a Date column and Sales, resample to monthly frequency and calculate the total sales.

```python
df['Join_Date'] = pd.date_range(start='2025-01-01', periods=len(df), freq='M')
df['Join_Date'].dt.year
df['Join_Date'].dt.month
```

```python
df.set_index('Join_Date', inplace=True)
df.resample('M').sum()
```
